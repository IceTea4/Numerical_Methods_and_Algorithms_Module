```python
import numpy as np
import matplotlib.pyplot as plt
```

Given function and its derivatives (sin(x) derivatives:
https://moodle.ktu.edu/mod/folder/view.php?id=434370 )

```python
def fx(x): # given function
  return np.sin(x)

# derivatives of different degree

def df1(x):
  return np.cos(x)

def df2(x):
  return -np.sin(x)

def df3(x):
  return -np.cos(x)

def df4(x):
  return np.sin(x)
```

Plotting given function

```python
dx= 0.1
x=np.arange(-10, 10+dx, dx)
y = fx(x)
plt.plot(x, y, 'b')
plt.grid()
```

Taylor series (different amount of terms)

```python
def ts0(x, x0):
  return fx(x0)

def ts1(x, x0):
  return fx(x0) + (x-x0)*df1(x0)

def ts2(x, x0):
  return fx(x0) + (x-x0)*df1(x0) + (np.power((x-x0), 2) / 2 ) *df2(x0)

def ts3(x, x0):
  return fx(x0) + (x-x0)*df1(x0) + (np.power((x-x0), 2) / 2 ) *df2(x0) + (np.power((x-x0), 3) / (3*2)) *df3(x0)

def ts4(x, x0):
  return fx(x0) + (x-x0)*df1(x0) + (np.power((x-x0), 2) / 2 ) *df2(x0) + (np.power((x-x0), 3) / (3*2)) *df3(x0) + (np.power((x-x0), 4) / (4*3*2)) *df4(x0)
```

```python
plt.xlim([-10, 10]); plt.ylim([-10, 10])

x0 = 2; # given point


# values of Taylor Series
tsy = ts4(x, x0)

plt.plot(x, y, 'b',  linewidth=3.0) # representing given function
plt.plot(x, tsy, 'r', 1) # representing given function
plt.plot(x0, fx(x0), 'or') # representing reference point

plt.legend(['Given func: f(x) = sin(x)', 'Taylor series (4th order)', 'Reference point'])
plt.grid()
```

```python
plt.xlim([-10, 10]); plt.ylim([-10, 10])

x0 = 2; # given point


# values of Taylor Series
tsy = ts3(x, x0)

plt.plot(x, y, 'b',  linewidth=3.0) # representing given function
plt.plot(x, tsy, 'r', 1) # representing given function
plt.plot(x0, fx(x0), 'or') # representing reference point

plt.legend(['Given func: f(x) = sin(x)', 'Taylor series (1th order)', 'Reference point'])
plt.grid()
```
