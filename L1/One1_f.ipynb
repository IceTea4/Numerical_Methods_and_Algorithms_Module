```python
import numpy as np
import math
import matplotlib.pyplot as plt

def func(x):
    return -1.29 * x**4 + 5.08 * x**3 - 2.76 * x**2 - 6.31 * x + 4.10

grubusRteig = 1 + 6.31 / 1.29
grubusRneig = -1 * grubusRteig

Bteig = 5.08
kteig = 4 - 3
tikslesnisRteig = 1 + (Bteig / 1.29)**(1/kteig)

Bneig = 6.31
kneig = 4 - 1
tikslesnisRneig = -1 * (1 + (Bneig / 1.29)**(1/kneig))

pi=math.pi
xxx=np.linspace(-2*pi, 2*pi, 100);
fff=func(xxx);

plt.figure(figsize=(12, 6), dpi=150)
plt.plot(xxx,fff,'b-');plt.grid();
plt.axhline(0, color='red', linestyle='--')
plt.ylim([-2700, 50])
plt.xlim([-6, 6])

plt.plot(grubusRteig, func(grubusRteig), 'ro')
plt.plot(grubusRneig, func(grubusRneig), 'ro')
plt.plot(tikslesnisRteig, func(tikslesnisRteig), 'go')
plt.plot(tikslesnisRneig, func(tikslesnisRneig), 'go')

print(f"Grubus: [{grubusRneig}; {grubusRteig}] (raudonas)")
print(f"Tikslesnis: [{tikslesnisRneig}; {tikslesnisRteig}] (Å¾alias)")
```
