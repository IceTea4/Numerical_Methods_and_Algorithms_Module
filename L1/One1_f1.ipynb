```python
import numpy as np
import math
import matplotlib.pyplot as plt

def func(x):
    return -1.29 * x**4 + 5.08 * x**3 - 2.76 * x**2 - 6.31 * x + 4.10

pi=math.pi
xxx=np.linspace(-2*pi, 2*pi, 100);
fff=func(xxx);

plt.figure(figsize=(12, 6), dpi=150)
plt.plot(xxx,fff,'b-');plt.grid();
plt.axhline(0, color='red', linestyle='--')
plt.ylim([-7, 7])
plt.xlim([-2, 4])
```
