```python
import numpy as np
import math
import matplotlib.pyplot as plt

def func(x):
    return x * (np.cos(x))**2 - (x / 2)**2

xneig = -10
xteig = 10

pi=math.pi
xxx=np.linspace(-11, 11, 100);
fff=func(xxx);

plt.figure(figsize=(12, 6), dpi=150)
plt.plot(xxx,fff,'b-');plt.grid();
plt.axhline(0, color='red', linestyle='--')
plt.ylim([-35, 5])
plt.xlim([-11, 11])

plt.plot(xteig, func(xteig), 'ro')
plt.plot(xneig, func(xneig), 'ro')

print(f"Duotas šaknų intervalas: [{xneig}; {xteig}]")
```
