```python
import numpy as np
import math
import sympy
from sympy import plot

x, f, fp, df = sympy.symbols(('x', 'f', 'fp', 'df'))
f = sympy.cos(2 * x)

x0 = 0
N = 5

fp = f.subs(x, x0)
fa = f

for i in range (1, N + 1):
    f = f.diff(x)
    fp = fp + f.subs(x, x0) / math.factorial(i) * (x - x0)**i
    
print(fp)

a = sympy.Poly(fp, x)
kf = np.array(a.all_coeffs())
saknys = np.roots(kf)
print(saknys)

p = plot(fa, fp, show = False)
p.ylim = (-30, 30)
p.xlim = (0, 4)
p.show()
```
