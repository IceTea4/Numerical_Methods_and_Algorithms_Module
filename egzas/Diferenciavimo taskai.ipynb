```python
import sympy as sp

# Išvalymas (analogiška clc, clear all, close all neturi tiesioginio atitikmens)
sp.init_printing()

# Simboliai
dx, x, a = sp.symbols('dx x a')

N = 9  # didžiausias diferencijavimo formulių taškų skaičius

for i in range(3, N + 1, 2):
    print(f"******************************    N={i}")

    # Taškai: a, a+dx, a+2dx, ...
    xx = [a + k * dx for k in range(i)]

    dL = []

    for j in range(i):
        # Lagranžo daugianaris
        L = 1
        for k in range(i):
            if k != j:
                L *= (x - xx[k]) / (xx[j] - xx[k])
        dL.append(sp.diff(L, x))

    print("\nDiferencijavimo formulės:")
    for j in range(i):
        DL = [dL[k].subs(x, a + j * dx) for k in range(i)]

        # Kaip MATLAB: DL * ((i-1)*dx) / ((i-1)*dx)
        expr = sp.Matrix(DL) * ((i - 1) * dx)
        print(expr, f"/({i-1}*dx)")

    print()
```
