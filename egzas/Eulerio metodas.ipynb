```python
import math

# Diferencialinės lygties dešinė pusė
def f(x, y):
    return -5+y*2 * math.sin(x)

# Pradinės sąlygos
x0 = 0.0
y0 = 0.0
h = 0.1
x_end = 0.2

# Eulerio metodas
x = x0
y = y0

while x < x_end - 1e-12:
    y = y + h * f(x, y)
    x = x + h

# Rezultatas
print(f"y({x_end}) ≈ {y}")
```
