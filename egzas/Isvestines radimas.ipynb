```python
import numpy as np

def diferencijavimas(f, dx, N):
    """
    Skaitinė išvestinė naudojant N taškų formules
    f  – funkcijos reikšmės masyvas
    dx – žingsnis
    N  – taškų skaičius (3,5,7,9)
    """

    koef = {
        3: np.array([
            [-3,  4, -1],
            [-1,  0,  1],
            [ 1, -4,  3]
        ]),

        5: np.array([
            [-25/3, 16, -12, 16/3, -1],
            [-1, -10/3, 6, -2, 1/3],
            [1/3, -8/3, 0, 8/3, -1/3],
            [-1/3, 2, -6, 10/3, 1],
            [1, -16/3, 12, -16, 25/3]
        ]),

        7: np.array([
            [-147/10, 36, -45, 40, -45/2, 36/5, -1],
            [-1, -77/10, 15, -10, 5, -3/2, 1/5],
            [1/5, -12/5, -7/2, 8, -3, 4/5, -1/10],
            [-1/10, 9/10, -9/2, 0, 9/2, -9/10, 1/10],
            [1/10, -4/5, 3, -8, 7/2, 12/5, -1/5],
            [-1/5, 3/2, -5, 10, -15, 77/10, 1],
            [1, -36/5, 45/2, -40, 45, -36, 147/10]
        ]),

        9: np.array([
            [-761/35, 64, -112, 448/3, -140, 448/5, -112/3, 64/7, -1],
            [-1, -446/35, 28, -28, 70/3, -14, 28/5, -4/3, 1/7],
            [1/7, -16/7, -38/5, 16, -10, 16/3, -2, 16/35, -1/21],
            [-1/21, 4/7, -4, -18/5, 10, -4, 4/3, -2/7, 1/35],
            [1/35, -32/105, 8/5, -32/5, 0, 32/5, -8/5, 32/105, -1/35],
            [-1/35, 2/7, -4/3, 4, -10, 18/5, 4, -4/7, 1/21],
            [1/21, -16/35, 2, -16/3, 10, -16, 38/5, 16/7, -1/7],
            [-1/7, 4/3, -28/5, 14, -70/3, 28, -28, 446/35, 1],
            [1, -64/7, 112/3, -448/5, 140, -448/3, 112, -64, 761/35]
        ])
    }

    if N not in koef:
        raise ValueError("N turi būti 3, 5, 7 arba 9")

    w = koef[N]
    n = len(f)
    df = np.zeros(n)
    mid = (N - 1) // 2

    for i in range(n):
        if i < mid:
            df[i] = np.dot(w[i], f[:N])
        elif i >= n - mid:
            df[i] = np.dot(w[N - (n - i)], f[-N:])
        else:
            df[i] = np.dot(w[mid], f[i-mid:i+mid+1])

    return df / ((N - 1) * dx)

def f(x):
    return np.sin(2*x - 2) + np.cos(5*x) + 1

a, b = -5, 5
n = 1000
x = np.linspace(a, b, n)
dx = x[1] - x[0]

fx = f(x)

df_num = diferencijavimas(fx, dx, N=7)
df_exact = 2*np.cos(2*x - 2) - 5*np.sin(5*x)

print("diff =", df_num[0])
print("diff real =", df_exact[0])
```
