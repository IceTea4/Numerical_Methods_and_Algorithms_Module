```python
# interpoliavimas 2D Lagranzo

import numpy as np
import matplotlib.pyplot as plt

#----------------------------------------------------------------------------
def Lagrange(X,n,x):
    N=len(X); L=np.ones(x.size,dtype=float)
    for i in range (N):
      if i != n:  L=L*(x-X[i])/(X[n]-X[i]);

    return L
 
#----------------------------------------------------------------------------

fig=plt.figure(1);ax=fig.add_subplot(1,1,1);ax.set_xlabel('x');ax.set_ylabel('y');
X=np.array([0, 1, 3, 5, 6, 3, 0, 2],dtype=float);Y=np.array([3, 10, 15, 2, 1, 4, 8, 10 ],dtype=float);
T=np.array([0, 1, 2, 3, 4, 5, 6, 7],dtype=float)
N=len(X);
ax.plot(X,Y,'bo')
plt.grid()

ttt=np.linspace(T[0],T[N-1],100)
xxx=np.zeros(ttt.size,dtype=float)
yyy=np.zeros(ttt.size,dtype=float)
for i in range (N):
  xxx+=X[i]*Lagrange(T,i,ttt)
  yyy+=Y[i]*Lagrange(T,i,ttt)

ax.plot(xxx,yyy,'r-')
ax.set_title('Lagrange');
```
