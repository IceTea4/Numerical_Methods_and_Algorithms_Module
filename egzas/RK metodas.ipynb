```python
import math

# Dešinės pusės funkcija f(x, y)
def f(x, y):
    return -5*y**2 + math.sin(x) + 2

# RK4 metodas
def rk4(x0, y0, h, x_end):
    x = x0
    y = y0

    while x < x_end - 1e-12:
        k1 = f(x, y)
        k2 = f(x + h/2, y + h*k1/2)
        k3 = f(x + h/2, y + h*k2/2)
        k4 = f(x + h, y + h*k3)
        
        y = y + h*(k1 + 2*k2 + 2*k3 + k4)/6
        x = x + h
    
    return y

# Duomenys
x0 = 0.0
y0 = 0.0
h = 0.1
x_end = 0.3

# Skaičiavimas
y_result = rk4(x0, y0, h, x_end)

# Rezultatas
print(f"y ≈ {y_result}")
```
